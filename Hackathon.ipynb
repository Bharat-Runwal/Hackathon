```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
import urllib.request, re
```

### Seek Nanny

```python
URL = "https://www.seek.com.au/nanny-jobs"
page = requests.get(URL)
soup = BeautifulSoup(page.text, "html.parser")
results = soup.find_all('article')
title = []
desc = []
data_dict = {}
data = pd.DataFrame()
for x in results:
        single_page_url = "https://www.seek.com.au" + x.contents[0].div['href']
        request2 = urllib.request.Request(url = single_page_url)
        response2 = urllib.request.urlopen(request2)
        single_temp_html = response2.read().decode()
        single_temp_html = BeautifulSoup(single_temp_html, "lxml")

        title=single_temp_html.title.string
        
        description=single_temp_html.find(attrs={"data-automation": "jobDescription"})
        desc = description.find("li").get_text()+' '+description.find("p").get_text()
        data_dict[title]=desc    
```

### Seek.com Babysitter

```python
data = pd.DataFrame(data_dict.items(),columns=['Title', 'Description'])
```
